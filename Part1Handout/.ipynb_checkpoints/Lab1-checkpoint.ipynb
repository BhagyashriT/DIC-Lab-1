# Problem 1

```r
sales1<-c(12,14,16,29,30,45,19,20,16, 19, 34, 20)
sales2<-rpois(12,34)  # random numbers, Poisson distribution, mean at 34, 12 numbers
par(bg="cornsilk")
mean(sales1)
mean(sales2)
plot(sales1, col="blue", type="o", ylim=c(0,100), xlab="Month", ylab="Sales" )
title(main="Sales by Month")

lines(sales2, type="o", pch=22, lty=2, col="red")
grid(nx=NA, ny=NULL)
legend("topright", inset=.05, c("Sales1","Sales2"), fill=c("blue","red"), horiz=TRUE)
```

Discussion: This is a simple line plot indicating sales of two products over a period of 12 months. The sales of product1 are stored in sales1 vector. Similarly, the sales of product2 are stored in sales2 vector. sales2 vector is declared with manual hard coded entries. sales2 vector has been assigned random values using poisson distribution. Hence, the plot of sales2 vector varies with every run of the code. Because the poisson distrubution of sales2 is centered at 34, the datapoints of sales2 will fall close to 34. Hence, average sales of product2 over 12 months is grater than average sales1. 

# Problem 2

```r
sales<-read.table(file.choose(), header=T)
sales  # to verify that data has been read
barplot(as.matrix(sales), main="Sales Data", ylab= "Total",beside=T, col=rainbow(5))
par(mfrow=c(1,2))
hist(sales$Sales1, main = "Histogram", col = rainbow(8))
hist(sales$Sales2, main = "Histogram", col = rainbow(8))
```

Discussion: This is a simple bar graph of sales of two products over 12 months. The data is read from a text file and stored in a data frame. It can be observed from the bar plot that sales for product1  were highest during August and September.</br>
From the side-by-side histogram of sales1 and sales2, it is observed 35 occurs most frequently for sales1 and 45,55 valeus occur most freqently for sales2.

# Problem 3 

```r
fn<-boxplot(sales,col=c("orange","green"))$stats

text(1.45, fn[3,2], paste("Median =", fn[3,2]), adj=0, cex=.7)
text(0.45, fn[3,1],paste("Median =", fn[3,1]), adj=0, cex=.7)
grid(nx=NA, ny=NULL)
```

Discussion: This is a box plot for sales of two products over 12 months. It can be observed that fewer data points lie below the median for sales1. For sales2, the data points are somewhat evenly distributed about the median. By observing the height of box plots, it can be inferred that data points of sales1 have more variance than data points of sales2.

```r
fb1<-read.csv(file.choose())
aapl1<-read.csv(file.choose())
par(bg="cornsilk")
plot(aapl1$Adj.Close, col="blue", type="o", ylim=c(150,200), xlab="Days", ylab="Price")
lines(fb1$Adj.Close, type="o", pch=22, lty=2, col="red")
legend("topright", inset=.05, c("Apple","Facebook"), fill=c("blue","red"), horiz=TRUE)
hist(aapl1$Adj.Close, col=rainbow(8))
```

Discussion: This is a simple line plot indicating the share values of Apples(blue) and Facebook(red) over 60 days. It can be observed that mean share value of facebook is greater than mean share value of Apple during the period of 60 days. On the 50th day, facebook had an exception as its stock price fell below Apple. 
In the histogram, we have plotted frequencies of share values of Apple. It can be observed that share value 175 has occured maximum number of times for apple. The share value of Apple has never been 165.

```r
library("ggmap")
library("maptools")
library(maps)
api_key <- Sys.getenv("apiKey")
register_google(key = api_key) 
visited <- c("SFO", "Chennai", "London", "Melbourne", "Lima","Peru","Johannesbury", "SA")
ll.visited <- geocode(visited)
visit.x <- ll.visited$lon
visit.y <- ll.visited$lat
map("world", fill=TRUE, col="white", bg="lightblue", ylim=c(-60, 90), mar=c(0,0,0,0))
points(visit.x,visit.y, col="red", pch=36)

```
